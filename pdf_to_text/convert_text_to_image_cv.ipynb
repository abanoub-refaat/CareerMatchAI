```python
from pdf2image import convert_from_path
from PIL import Image
import os
```

```python
PDF_PATH = "text_based_cv.pdf"
OUTPUT_PATH = "image_based_cv.pdf"
```

### convert pdf_text_based_to_image
#### for test only

```python

def pdf_to_images_to_pdf(input_pdf_path, output_pdf_path=OUTPUT_PATH, dpi=300):
    """
    Convert a PDF to images and back to a combined PDF automatically.
    
    Args:
        input_pdf_path (str): Path to the input PDF.
        output_pdf_path (str): Path to save the output PDF.
        dpi (int): Resolution for converting PDF pages to images.
    """
    pages = convert_from_path(input_pdf_path, dpi=dpi)
    
    images = [page.convert("RGB") for page in pages]
    
    images[0].save(output_pdf_path, save_all=True, append_images=images[1:])
    
    print(f"✅ PDF created: {output_pdf_path}")

```

```python
pdf_to_images_to_pdf(PDF_PATH, OUTPUT_PATH)
```
